# Basic Linear Regression

# Get input data


```python
# Gettting 80% of the data for training and setting X to be the parameters and Y be the saleprice
df_train_80 = df_train[:per_80]
df_X = df_train_80.loc[:,['OverallQual','GrLivArea','GarageCars','TotalBsmtSF','FullBath', \
                     'YearBuilt']]
df_y = df_train_80.loc[:, ['SalePrice']]
```

# Data Preprocessing

```python
import torch
from torch.autograd import Variable
```

```python
# Switching Data Frames into Variables to be able to run Linear Regression
X = Variable(torch.Tensor(df_X.values.tolist()))
y = Variable(torch.Tensor(df_y.values.tolist()))
```

# Linear Regression

```python
total_houses = per_80 
total_parameters = 6
W = Variable(torch.ones(total_parameters, 1), requires_grad=True)
b = Variable(torch.ones(1), requires_grad=True)
```

```python
import torch.nn as nn

learning_rate = 1e-2
optimizer = torch.optim.Adam([W, b], lr=learning_rate)
loss_fxn = nn.MSELoss()
losses = []
t = 0
# Setting it to this since the highest loss was 31e6. 
prev_loss = 4e8
stillTraining = True
while stillTraining:
    predictions = X.mm(W) + b
    loss = loss_fxn(predictions, y)
    losses.append(loss.data[0] / total_houses)
    
    if ((prev_loss - (loss.data[0] / total_houses)) < 1e-2): 
        stillTraining = False
        
    prev_loss = loss.data[0] / total_houses 
    t = t + 1
    
    optimizer.zero_grad()   
    loss.backward()
    optimizer.step()
```

```python
# Plotting results from training
import matplotlib.pyplot as plt
plt.plot(losses)
plt.title('Training plot')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.legend()
plt.grid()
plt.show()
```

# Checking out graphs

```python
plt.scatter(predictions.data.numpy(), y.data.numpy(),  color='g')
plt.plot(y.data.numpy(), y.data.numpy(), color='b')
plt.title('Final Findings')
plt.xlabel('Predictions')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
df_train_20 = df_train[:per_20]
test_total_houses = per_20
test_df_X = df_train_20.loc[:,['OverallQual','GrLivArea','GarageCars','TotalBsmtSF','FullBath', \
                     'YearBuilt']]
test_df_y = df_train_20.loc[:, ['SalePrice']]
test_X = Variable(torch.Tensor(test_df_X.values.tolist()))
test_y = Variable(torch.Tensor(test_df_y.values.tolist()))

test_predictions = test_X.mm(W) + b
test_loss = loss_fxn(test_predictions, test_y)

plt.scatter(test_y.data.numpy(), test_predictions.data.numpy(), color='g')
plt.plot(test_y.data.numpy(), test_y.data.numpy(), color='b')
plt.title('Test')
plt.xlabel('Predictions')
plt.ylabel('y')
plt.legend()
plt.show()
```
